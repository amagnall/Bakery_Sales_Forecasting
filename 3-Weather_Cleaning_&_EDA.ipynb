# Wattisham Weather Dataset Cleaning

Source of this dataset: National Centres For Environmental Information

Data dictionary: 

- TMAX - Maximum temperature in Farenheight
- TAVG - Average Temperature in Farenheight
- TMIN - Minimum temperature in Farenheight
- PRCP - Precipitation, in inches
- SNWD - Snow depth, in inches

This dataset shows the daily weather from the Wattisham weather station from 01/07/2021 to 18/09/2022. 

## Imports

```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
import datetime as dt
import time
import os
```

```python
working_directory = os.getcwd()
print(working_directory)
```

```python
raw_weather = pd.read_csv(working_directory+'/1_raw_data/raw_weather_data.csv')
```

```python
raw_weather
```

## Data Exploration

```python
clean_weather = raw_weather[["DATE", "TAVG", "TMAX", "TMIN", "PRCP", "SNWD"]].copy()
```

```python
clean_weather.info()
```

```python
clean_weather.duplicated().sum()
```

```python
clean_weather.isna().sum()
```

## Observations

- Can see that TMax, min, prcp and snwd havce missing values. As only need `TAV` and there are no missing values can drop the `TMAX` and `TMIN` columns 
- Need to populate the `PRCP` and `SNWD` null columns with 0, assuming they are 0
- Date column is the wrong data type
- No duplicated rows
- `SNWD` only contains 11 rows of data, so should either be dropped or combined with another column 

```python
clean_weather = clean_weather.drop(columns=['TMAX', 'TMIN'])
clean_weather
```

## Data Cleaning

```python
clean_weather['DATE'] =  pd.to_datetime(clean_weather['DATE'], format='%d/%m/%Y')
```

```python
clean_weather['DATE'] = clean_weather['DATE'].astype('datetime64')
```

```python
# Validate it worked 
clean_weather.info()
```

```python
processed_weather = clean_weather.copy()
```

As the `SNWD` column only contains 11 rows of data, they will be added into the `PRCP` column as alone they will not provide insight as there is insufficient data to build a model of and for sales to be dependent on. 

```python
processed_weather['PRCP'].fillna(0,inplace=True)
processed_weather['SNWD'].fillna(0,inplace=True)
```

```python
processed_weather['PRCP'] = processed_weather['PRCP'] + processed_weather['SNWD']
```

```python
processed_weather.drop(columns=['SNWD'],inplace=True)
```

```python
# Validate cols filled and dropped
processed_weather.info()
```

```python
processed_weather.describe()
```

It can be seen that `TAVG` is in Fahrenheit so this will be converted to Celsius. 

```python
processed_weather['TAVG'] = processed_weather['TAVG']-32
```

```python
processed_weather.describe()
```

There appears to be an outlier TAVG of 52 degrees Celcius. To explore this further the distriubtion will be plotted as a histogram. 

```python
sns.histplot(data=processed_weather['TAVG'])
```

```python
hot_dates = processed_weather[processed_weather['TAVG']>35]
```

Know from additional research that the max temperature reached in Suffolk was c.38 Degrees Celcius (in 2022) and c.35 for 2021. Therefore, any above these values will be reduced to 35 degrees for 2021 and 28 for 2022. 

```python
hot_dates
```

```python
for row in hot_dates.index:
    if row in [17,18,20,68,69]:
        processed_weather.loc[row,'TAVG'] = 35
    else:
        if hot_dates.loc[row,'TAVG'] > 38:
            processed_weather.loc[row,'TAVG'] = 38
```

```python
processed_weather.describe()
```

Confirmed now the temp, rain and snow are all valid. 

```python
processed_weather.rename(columns={'DATE':'Date'},inplace=True)
processed_weather
```

## To Validate All Dates are Present in the Dataset
For timeseries analaysis there need to be rows of data for every date. Therefore to validate all dates are present.. 

```python
first_day = processed_weather['Date'].min()
last_day=processed_weather['Date'].max()
full_range = pd.date_range(start=first_day, end=last_day, freq="D")

# to identify the missing dates in the dataframe 
full_range.difference(processed_weather['Date'])
```

```python
processed_weather.info()
```

### Observations
Can see that no dates were returned in the first step, suggesting that all dates are present in the `Date` column. In addition to this, the length of the dataframe is 760 days which is the correct length of days between 01/09/2020 and 01/10/2022. Therefore, all the dates are present in the dataset, there are no null values, redundant columns of duplicated rows and the processed dataset can be saved. 

```python
processed_weather.to_csv('processed_weather.csv',index=False)
```
